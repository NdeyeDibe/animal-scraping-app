
**CHIEN**

```python
# url
url = 'https://sn.coinafrique.com/categorie/chiens?page=1'
# récupérer le contenu de la page
res= get(url)
# res.content # pour voir le contenu de la page
# Avoir le contenu dans un objet bs et bien sparsifier
soup = bs(res.content, 'html.parser')
```

```python
# récupérer les conteneurs
containers = soup.find_all('div', 'col s6 m4 l3')
len(containers)
```

```python
# premier container
container = containers[0]
# container
# Name chien
name_c = container.find('p', 'ad__card-description').a.text
# name_c
# price
price_c = container.find('p', 'ad__card-price').text.strip('CFA')
#price_c
# adress chien
adress_c = container.find('p', 'ad__card-location').span.text
#adress_c
# url image chien
url_image_c = container.find('img', 'ad__card-img')['src']
print(name_c,'--',price_c, '--',adress_c, '--', url_image_c)
```

```python
# récupérer les informations sur tous les containers
data = []
for container in containers:
  try:
    # Name chien
    name_c = container.find('p', 'ad__card-description').a.text
    # price chien
    price_c = container.find('p', 'ad__card-price').text.strip('CFA')
    # adress chien
    adress_c = container.find('p', 'ad__card-location').span.text
    # url image chien
    url_image_c = container.find('img', 'ad__card-img')['src']
    dic = {
        'name_c': name_c,
        'price_c': price_c,
        'adress_c': adress_c,
        'url_image_c':url_image_c
    }
    data.append(dic)
  except:
    pass

# Dataframe
DF1 = pd.DataFrame(data)
```

```python
DF1.head()
```

```python
DF1.shape
```

```python
# scraper les données sur plusieurs
df1_final = pd.DataFrame()
for ind_page in range(1, 12):
  # url
  url = 'https://sn.coinafrique.com/categorie/chiens?page={ind_page}'
  # récupérer le contenu de la page
  res= get(url)
  # res.content # pour voir le contenu de la page
  # Avoir le contenu dans un objet bs et bien sparsifier
  soup = bs(res.content, 'html.parser')
  # récupérer les conteneurs
  containers = soup.find_all('div', 'col s6 m4 l3')
  len(containers)

  data = []
  for container in containers:
    try:
      # Name chien
      name_c = container.find('p', 'ad__card-description').a.text
      # price chien
      price_c = container.find('p', 'ad__card-price').text.strip('CFA')
      # adress chien
      adress_c = container.find('p', 'ad__card-location').span.text
      # url image chien
      url_image_c = container.find('img', 'ad__card-img')['src']
      dic = {
          'name_c': name_c,
          'price_c': price_c,
          'adress_c': adress_c,
          'url_image_c':url_image_c
      }
      data.append(dic)
    except:
      pass

  # Dataframe
  DF1 = pd.DataFrame(data)
  df1_final = pd.concat([df1_final, DF1], axis = 0).reset_index(drop = True)
```
